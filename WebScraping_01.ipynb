Importing the libraries "requests" and "BeautifulSoup"

```python
import requests
from bs4 import BeautifulSoup
```

Passing the URL into a variable

```python
url = "http://wikipedia.org"
```

Using the request library to fetch the URL

```python
r= requests.get(url)
```

```python
htmlContent = r.content
```

Creating a BeautifulSoup object

```python
soup = BeautifulSoup(htmlContent, 'html.parser')
print(soup.prettify())
```

Creating a list of heading tags()

```python
heading_tags = ["h1","h2","h3","h4","h5","h6"]
```

Iterating over all the heading tags using find_all() method

```python
for tags in soup.find_all(heading_tags):
    print(tags.name + '->' + tags.text.strip())
```
